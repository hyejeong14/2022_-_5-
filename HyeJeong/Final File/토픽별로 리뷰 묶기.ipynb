```python
df2=pd.read_csv('all_review_.csv')
df2
```

```python
content=df2['CONTENT']
```

```python
df['문서 번호']=content
df=df.drop([df.columns[0]],axis=1)
df
```

```python
df.insert(0,'별점',df2['STAR'])
```

```python
df
```

```python
df.rename(columns={'문서 번호':'리뷰 내용','가장 비중이 높은 토픽':'토픽'},inplace=True)
df
```

```python
topic1=pd.DataFrame(df[df['토픽']==0]['리뷰 내용'])
star=pd.DataFrame(df[df['토픽']==0]['별점'])
topic1=pd.concat([star,topic1],axis=1)
topic1
```

```python
topic2=pd.DataFrame(df[df['토픽']==1]['리뷰 내용'])
star=pd.DataFrame(df[df['토픽']==1]['별점'])
topic2=pd.concat([star,topic2],axis=1)
topic2
```

```python
topic3=pd.DataFrame(df[df['토픽']==2]['리뷰 내용'])
star=pd.DataFrame(df[df['토픽']==2]['별점'])
topic3=pd.concat([star,topic3],axis=1)
topic3
```

```python
topic4=pd.DataFrame(df[df['토픽']==3]['리뷰 내용'])
star=pd.DataFrame(df[df['토픽']==3]['별점'])
topic4=pd.concat([star,topic4],axis=1)
topic4
```

```python
topic1.to_csv("토픽1.csv")
topic2.to_csv("토픽2.csv")
topic3.to_csv("토픽3.csv")
topic4.to_csv("토픽4.csv")
```

```python
TR=pd.DataFrame([topic1,topic2,topic3,topic4],index=['Topic 1','Topic 2','Topic 3','Topic 4']).transpose()
TR.to_csv("토픽 별 리뷰 1.csv")
    
```
