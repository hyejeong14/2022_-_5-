```python
import pandas as pd
import numpy as np
import re
df = pd.read_csv('War Memorial of Korea_tripadvisor.csv')
df
```

```python
df.isnull().sum()
```

```python
df['REVIEW'].replace('', np.nan, inplace=True)
df
```

```python
df.isnull().sum()
```

```python
 emoji = re.compile("["
                    u"\U0001F600-\U0001F64F"
                    u"\U0001F300-\U0001F5FF"
                    u"\U0001F680-\U0001F6FF"
                    u"\U0001F1E0-\U0001F1FF"
                       "]+", flags=re.UNICODE)
        
        #이모티콘,특수문자, 한글 제거
content_noemj=[]
name_noko=[]
        
korean = re.compile('[\u3131-\u3163\uac00-\ud7a3]+')

for i in df['NAME'] :
    i = emoji.sub(r'',i)
    i = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\₩[\]\<\>`\'…《\》’]', '', i)
    i = re.sub(korean,'', i)
    i = i.encode('utf-8').decode('ascii','backslashreplace')
    name_noko.append(i)
df['NAME'] = name_noko            
for i in df['REVIEW'] :
    i = emoji.sub(r'',i)
    i = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\₩[\]\<\>`\'…《\》’]', '', i)
    i = re.sub(korean,'', i) 
    i = i.encode('utf-8').decode('ascii','backslashreplace')
    content_noemj.append(i)
df['REVIEW'] = content_noemj

df
```

```python
df.isnull().sum()
```

```python
df.to_csv('war memorial of korea_trip.csv')
```
