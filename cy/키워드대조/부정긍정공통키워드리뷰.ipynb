```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
senti_analyzer = SentimentIntensityAnalyzer()
list_pos=[]
list_neg=[]

for i in fk.index:
    senti_scores = senti_analyzer.polarity_scores(fk['CONTENT'][i])
    if senti_scores['compound'] >0.05 and fk['STAR'][i] >= 4 :
        list_pos.append(fk['CONTENT'][i])
        #print()
        #print(len(list_pos))
    
    elif senti_scores['compound']<= -0.05 or fk['STAR'][i] <4:
        list_neg.append(fk['CONTENT'][i])
        #print()
        #print(len(list_neg))
    #print(senti_scores, " ",i)
print(len(list_pos))
print(len(list_neg))
```

```python
import matplotlib.pyplot as plt
value=[len(list_pos),len(list_neg)]
group_names=['Positive Reviews','Negative Riveiews']
group_size=[3642,2714]
explode=[0.05,0.05]
colors=['lightgreen','gold']
plt.pie(value,labels=group_names,autopct='%.1f%%',startangle=360,explode=explode,shadow=True,colors=colors)
plt.axis('equal')
plt.title('Area',fontweight= 'bold',fontsize='x-large')
plt.savefig('area__piechart.png')
```
