```python
import nltk
import pandas as pd
import re
import numpy as np
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

df1=pd.read_csv('Shinsegye.csv').dropna().reset_index()
df1
```

```python
emoji = re.compile("["
                    u"\U0001F600-\U0001F64F"
                    u"\U0001F300-\U0001F5FF"
                    u"\U0001F680-\U0001F6FF"
                    u"\U0001F1E0-\U0001F1FF"
                       "]+", flags=re.UNICODE)
def trans_str(st):
    if st.startswith('('):
        table=str.maketrans(')(','  ')
        m=st.translate(table)
        #m1=re.sub(r'[^a-zA-Z0-9]', '', m)
       # m1= emoji.sub(r'',m1)
        #m1= m1.encode('utf-8').decode('ascii','ignore')
        result=re.search('Translated by Google(.+?)Original', m).group(1)
    else: result=st
        
    return result

def emo(emo):
    emo= emoji.sub(r'',emo)
    emp= emo.encode('utf-8').decode('ascii','ignore')
    
    return emo

def dro(test):
    if test in 'kkk':
        test=np.nan
        
    return test
```

```python

df1['리뷰']=df1['리뷰'].apply(trans_str)
df1['리뷰']=df1['리뷰'].apply(emo)
df1['리뷰'] = [re.sub('[^A-Za-z0-9 ]','', s) for s in df1['리뷰']]
```

```python

for i in range(len(df1)):
    word = df1['리뷰'][i].split()
    #df1['리뷰'][i]=word
    clean_review = ' '.join(word)
    df1['리뷰'][i]=clean_review
        
df1['리뷰']= df1['리뷰'].replace([''],[np.nan])
```

```python
df1=df1.dropna()
```
