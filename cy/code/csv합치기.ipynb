```python
import pandas as pd
```

```python
df_goY=pd.read_csv('go_Yeouido.csv')
df_goBuk=pd.read_csv('go_Bukchon.csv')
df_goSin=pd.read_csv('go_Shinsegye.csv')

df_goLot=pd.read_csv("final Lotte World Aquarium review.csv")
df_goMyeong=pd.read_csv('final Myeong Dong Walking Street review.csv')
df_goStar=pd.read_csv('final Starfield COEX Mall review.csv')

df_Lot=pd.read_csv('Lotte World Aquarium_tripadvisor.csv')
df_Myeong=pd.read_csv('Myeong Dong Walking Street_tripadvisor.csv')
df_Star=pd.read_csv('Starfield COEX Mall_tripadvisor.csv')


df_goDeok=pd.read_csv('final Deoksugung review.csv')
df_goNat=pd.read_csv('final National Museum of Korea review.csv')
df_goWar=pd.read_csv('final War Memorial Korea review.csv')
df_Deok=pd.read_csv('deoksugung_trip.csv')
df_Nat=pd.read_csv('national museum of korea_trip.csv')
df_War=pd.read_csv('war memorial of korea_trip.csv')
```

```python
# 서진 혜정 구글 리뷰 합치기
df_goLMS=pd.concat([df_goLot,df_goMyeong,df_goStar])
df_goDNW=pd.concat([df_goDeok,df_goNat,df_goWar])
df_LD=pd.concat([df_goLMS,df_goDNW])
#열 이름 수정
re_LD=df_LD[['STAR','CONTENT','DATE']]
re_LD=re_LD.rename(columns={'CONTENT':'REVIEW'})
```

```python
re_LD
```

```python
#혜정 서진 트립 리뷰 합치기
df_tripLMS=pd.concat([df_Lot ,df_Myeong,df_Star])
df_tripDNW=pd.concat([df_Deok ,df_Nat,df_War])
print(df_tripLMS.columns)
print(df_tripDNW.columns)
#print(df_tripDNW.columns)
re_tripLMS=df_tripLMS[['별점','리뷰','날짜']]
re_tripDNW=df_tripDNW[['STAR','REVIEW','DATE']]

re_tripLMS=re_tripLMS.rename(columns={'별점':'STAR','리뷰':'REVIEW','날짜':'DATE'})
print(re_tripLMS.columns)
print(re_tripDNW.columns)
retrip_LD=pd.concat([re_tripLMS,re_tripDNW])
```

```python
retrip_LD
```

```python
sum_HS=pd.concat([re_LD,retrip_LD])
```
