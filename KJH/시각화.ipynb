```python
def reviewplot(filename):
    plt.figure(figsize=(30,10))
    X=[]
    Y=[]
    for file in filename:
        a=pd.read_csv(file)
        X.append(file.split('.csv')[0].split('다시 크롤링/')[1])
        Y.append(a.shape[0])
    plt.bar(X,Y)
    plt.ylabel('Number of Reviews',fontsize=18)
    plt.yticks(fontsize=20)
    plt.xticks(fontsize=20,rotation=90)
    for i, v in enumerate(X):
        plt.text(v, Y[i], Y[i],                 # 좌표 (x축 = v, y축 = y[0]..y[1], 표시 = y[0]..y[1])
             fontsize = 20,
             horizontalalignment='center',  # horizontalalignment (left, center, right)
             verticalalignment='bottom')    # verticalalignment (top, center, bottom)
    plt.show()
    print(sum(Y))
```

```python
def starplot(filename):
    plt.figure(figsize=(30,10))
    X=[]
    Y=[]
    for file in filename:
        a=pd.read_csv(file)
        X.append(file.split('.csv')[0].split('다시 크롤링/')[1])
        Y.append(a['STAR'].mean())
    plt.bar(X,Y)
    plt.ylabel('Number of Reviews',fontsize=18)
    plt.yticks(fontsize=20)
    plt.xticks(fontsize=20,rotation=90)
    for i, v in enumerate(X):
        plt.text(v, Y[i], round(Y[i],2),                 # 좌표 (x축 = v, y축 = y[0]..y[1], 표시 = y[0]..y[1])
             fontsize = 20,
             horizontalalignment='center',  # horizontalalignment (left, center, right)
             verticalalignment='bottom')    # verticalalignment (top, center, bottom)
    plt.show()
    print(sum(Y))
```

```python
#--------------------------------------------------------------긍부정 비율 시각화------------------------------------------------
```

```python
def posnegpiechart(filename):
    for file in filename:
        X=[]
        Y=[]
        a=pd.read_csv(file)
        num_pos=sum(a['STAR']>=4.0)
        num_neg=sum(a['STAR']<4.0)
        ratio=[num_pos,num_neg]
        label=['Positive','Negative']
        plt.title(file.split('.csv')[0].split('다시 크롤링/')[1])
        plt.pie(ratio,labels=label,autopct='%.1f%%')
        plt.show()
```
