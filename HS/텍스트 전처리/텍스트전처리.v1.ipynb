토큰화 & 불용어 제거

```python
def text_clearing(text):
    text = text.lower()
    text = re.sub('[^ a-z A-Z 0-9]+',' ',str(text))
    text = [word.strip(string.punctuation) for word in text.split()]
    stop = stopwords.words('english')
    text = [x for x in text if x not in stop]
    text = [t for t in text if len(t)>0]
    pos_tags = pos_tag(text)
    text = [t for t in text if len(t)>1]
    text = " ".join(text)
    return(text)
```

```python
df["CONTENT"]=df["CONTENT"].apply(lambda x:text_clearing(x))
df
```

어간(Stemming) 및 표제어(Lemmatization)

```python
# Stemming
stemmer = PorterStemmer()
def text_stemming(text):
    text = text.lower()
    text = str(text)
    text = text.split()
    stem_words = []
    for w in text:
        x = stemmer.stem(w)
        stem_words.append(x)
    stem_words = " ".join(stem_words)
    return(stem_words)
```

```python
df["STEM"]=df["CONTENT"].apply(lambda x:text_stemming(x))
df
```

```python
# Lemmatization
lemma = WordNetLemmatizer()
def text_lemmatizer(text):
    text = text.lower()
    text = str(text)
    text = text.split()
    lemmatizer_words = []
    for w in text:
        x = lemma.lemmatize(w)
        lemmatizer_words.append(x)
    lemmatizer_words = " ".join(lemmatizer_words)
    return(lemmatizer_words)
```

```python
df["LEMMA"]=df["CONTENT"].apply(lambda x:text_lemmatizer(x))
df
```
