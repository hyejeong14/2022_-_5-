```python
import nltk
import string
import pandas as pd
import numpy as np
import contractions
from nltk.corpus import stopwords, wordnet
from nltk.tokenize import word_tokenize
from autocorrect import Speller
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from nltk.stem.wordnet import WordNetLemmatizer
```

Read Document

```python
df = pd.read_csv('Dongdaemun Design Plaza.csv')
df
```

```python
for col in df.columns:
    print(col,df[col].isnull().sum())
```

```python
rws = df.loc[:,['STAR','CONTENT']]
```

Tokenize

```python
df['tokenized'] = df['CONTENT'].apply(word_tokenize)
df.head()
```

대소문자 통합

```python
df['lower'] = df['tokenized'].apply(lambda x: [word.lower() for word in x])
df.head()
```

불용어, 불필요 단어 제거

```python
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
df['stopwords_removed'] = df['lower'].apply(lambda x : [word for word in x if word not in stop_words])
df.head()
```

어간 추출, 기본형 추출

```python
nltk.download('averaged_perceptron_tagger')
df['pos_tags'] = df['stopwords_removed'].apply(nltk.tag.pos_tag)
df.head()
```

```python
nltk.download('wordnet')
nltk.download('omw-1.4')
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
    
df['wordnet_pos'] = df['pos_tags'].apply(lambda x: [(word, get_wordnet_pos(pos_tag)) for (word,pos_tag)in x])
df
```

```python
df.to_csv('DDP_전처리.csv')
```
